### Analysis the Categorical Data Distribution

#### Crop
- Rice and Maize appears the most number of times in the dataset

```python
plt.figure(figsize = (10,9))
sns.countplot(y="Crop",data=df, order=df['Crop'].value_counts().index)
plt.title("Crop Distribution")
plt.show()
```

#### Season
- Kharif appears the most number of times in the given data set

```python
plt.figure(figsize = (5,5))
sns.countplot(y="Season",data=df, order=df['Season'].value_counts().index)
plt.title("Season Distribution")
plt.show()
```

#### State
- Karnataka appears the most number of times in the given dataset 
- Delhi appears the least number of times in the given dataset

```python
plt.figure(figsize = (10,9))
sns.countplot(y="State",data=df, order=df['State'].value_counts().index)
plt.title("State Distribution")
plt.show()
```

```python
df[['Area', 'Production', 'Annual_Rainfall', 'Fertilizer', 'Pesticide', 'Yield']].hist(bins=20, figsize=(12, 10))
plt.suptitle("Histograms for Numerical Columns")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df[['Area', 'Production', 'Annual_Rainfall', 'Fertilizer', 'Pesticide', 'Yield']])
plt.title("Boxplots for Numerical Columns")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,12))
sns.scatterplot(x='Area', y='Production', hue='Crop', data=df)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.tight_layout()
plt.title("Production vs Area")
plt.show()
```

```python
plt.figure(figsize=(10,12))
sns.scatterplot(x='Fertilizer', y='Yield', hue='Crop', data=df)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.tight_layout()
plt.title("Yield vs Fertilizer ")
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='Annual_Rainfall', y='Production', hue='State', data=df)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.tight_layout()
plt.title("Production vs Annual Rain by State")
plt.show()
```

```python
plt.figure(figsize=(10,12))
sns.lineplot(x='Crop_Year', y='Yield', hue='Crop', data=df)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize='small')
plt.tight_layout()
plt.title("Yield Trends Over the Years by Crop")
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
sns.scatterplot(x="Area", y="Yield", data=df, ax=axes[0, 0])
sns.scatterplot(x="Production", y="Yield", data=df, ax=axes[0, 1])
sns.scatterplot(x="Annual_Rainfall", y="Yield", data=df, ax=axes[0, 2])
sns.scatterplot(x="Fertilizer", y="Yield", data=df, ax=axes[1, 0])
sns.scatterplot(x="Pesticide", y="Yield", data=df, ax=axes[1, 1])
plt.suptitle("Scatter Plots of Yield vs Numerical Features")
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(x="Crop", y="Yield", data=df)
plt.xticks(rotation=90)
plt.title("Yield Distribution by Crop")
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(x="Season", y="Yield", data=df)
plt.title("Yield Distribution by Season")
plt.show()
```
